Each goal has several targets and indicators that are quantifiable metrics. The dataset used in this project gathered from government agencies in the various countries from 2000 to 2015 to support Goal 3: Good Health and Well-being. 

#### This dataset has been preprocessed, reshaped using Python and rows and columns with missing data are removed by Alan Fekete. So, we only performed exploratory data analysis with this simplified data.

# Import Data Science Libraries

```python
#import data science libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# loading the data with new names
data = pd.read_csv("SDG_goal3_clean.csv", names=["con", "yr", "reg", "mat_mort", "bir_hlth", "inf_mortU", "inf_mortM", "inf_mortF", "u5_mortU", "u5_mortM", "u5_mortF", "neo_mort", "mort_disU", "mort_disM", "mort_disF", "diab_dth", "cancr_dth", "sui_mortU", "sui_mortM", "sui_mortF", "rd_dthU", "rd_dthM", "rd_dthF", "adol_bth", "uhc", "phy_wrk", "nurs_wrk", "pham_wrk"])
# previewing the dataset's first 2 rows
data.head(2)
```

```python
#removing coloums and creating a new dataframe
df1=data.drop(0)
df2 = df1.drop(columns=df1.columns[3:12])
df = df2.drop(columns=df2.columns[6:15])
df.head(2)
```

# Mortality Rate and Healthcare System

#### Healthcare infrastructure (number of physicians, nurses, pharmacists) is closely connected to mortality rate of a country. Grouping them into qualitative variables gives a better understanding of how healthcare can affect lifespan of the citizens.

```python
#avg number of physicans, nurses, pharmacists
df['avg_healthcare'] = df[['phy_wrk', 'nurs_wrk', 'pham_wrk']].astype(float).sum(1)/3
df.head(2)
```

```python
# grouping data with eqally distributed bins based on avg number of physicans, nurses, pharmacists
df['avg_healthcare'] = df['avg_healthcare'].astype(float)

min_value = df['avg_healthcare'].min()
max_value = df['avg_healthcare'].max()
print(min_value)
print(max_value)

bins = [0.5,35,60,90]

healthcare_rating = ['Poor Healthcare','Moderate Healthcare', 'Good Healthcare']
df['hlt_rating'] = pd.cut(df['avg_healthcare'], bins=bins, labels=healthcare_rating, include_lowest=True)

#display the new grouped dataframe
df.head(2)
```

```python
#grouped aggregates for the grouped dataframe
df.groupby('hlt_rating')['mort_disU'].count()
```

#### Below is a grouped aggregate that shows the amount of countries with varying healthcare ratings from 2000-2015. We can see that there are a high amount of deaths (91) recorded due to poor healtcare services. I used region group aggregate and do a simple pie chart visualize mortalty rate by regions present in the dataset. Countries with poor healtcare facilities has higher mortality rate (56%) while citizens with the good facilities are seen to have higher lifespan. 

#### Pie chart is more compactly designed and the proportions are easier to spot out compared to a bar graph, hence the decision to graph the regional breakdown using pie chart has been made.

```python
#pie chart showing regional mortality rate of the countries available in spreadsheet
region_sums = df.groupby('hlt_rating')['reg'].count()
plt.axis('equal')
plt.pie(region_sums, labels = region_sums.index, autopct = '%1.0f%%')

#Title and legend
plt.title("Region Breakdown")
plt.legend()

#Display
plt.tight_layout()
plt.show()
```

```python
# bar plot with standard deviation
df['mort_disM'] = df['mort_disM'].astype(float)
df['mort_disF'] = df['mort_disF'].astype(float)

X1 = df.groupby('reg')['mort_disM'].mean()
X2 = df.groupby('reg')['mort_disF'].mean()

regions = df.groupby('reg')['mort_disF'].count()
y_labels = regions.index

fig = plt.figure(figsize=(10,6))

# plot bars
y_pos = np.arange(len(X1))
y_pos = [x for x in y_pos]
plt.yticks(y_pos, y_labels, fontsize=10)

# plot bars
y_pos = np.arange(len(X1))
y_pos = [x for x in y_pos]
plt.yticks(y_pos, y_labels, fontsize=10)

plt.barh(y_pos, X1, align='center', alpha=0.5, color='g')

# the second bar:
plt.barh(y_pos, -X2, align='center', alpha=0.5, color='b')

# annotation and labels
plt.xlabel('Mortality Rate')
t = plt.title('Male vs Female Regional Mortality Rate')
plt.ylim([-1,len(X1)+0.1])
plt.xlim([-max(X2)-1, max(X1)+1])
plt.grid()
plt.legend(['Male', 'Female'], loc='upper right')

plt.show()
```

#### Having a look at both pie chart and bar plot, we can conclude that there is quite a strong relationship between death rate by various disease and number of healthcare professionals. It is evident that male death rate is higher than female in all the regions despite of having higher number of healthcare professionals.
